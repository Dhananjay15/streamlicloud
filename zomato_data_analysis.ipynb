# **DATA CLEANING**

**On the basis of above analysis we can further move on to data cleaning process**

**Deleting redundant columns**

```python
redundant_columns=['url','address','phone','dish_liked','menu_item']
df.drop(redundant_columns,inplace=True,axis=1)
```

Above features are considered redundant due the amount of missing values or not enough importnace in further analysis

**Renaming the columns**

```python
df.rename(columns={'approx_cost(for two people)':'approx_cost_for_two'}, inplace=True)
```

Don't think there is major need to do that, still did for one.

**Dropping duplicates**

There are no duplicate entries present

**Cleaning individual columns and Remove the NaN values from the dataset**

```python
#removing unnecessary characters
characters_to_remove = ['/5',' /5','NEW','-']
for i in characters_to_remove:
    df['rate'] = df['rate'].str.replace(i,'')

df['rate']
```

```python
#replacing nan values with zero
df['rate'] = df['rate'].fillna(0)

#converting data type to float
df['rate'] = pd.to_numeric(df["rate"], downcast="float")

#replacing zero values with mean of column
df['rate'].replace(to_replace = 0, value = df['rate'].mean(), inplace=True)

df['rate']
```

```python
#same for approx_cost_for_two
characters_to_remove = [',']
for i in characters_to_remove:
    df['approx_cost_for_two'] = df['approx_cost_for_two'].str.replace(i,'')
    
df['approx_cost_for_two']
```

```python
df['approx_cost_for_two'] = df['approx_cost_for_two'].fillna(0)
df['approx_cost_for_two'] = pd.to_numeric(df["approx_cost_for_two"], downcast="integer")
```

```python
df['approx_cost_for_two'].describe()
```

value ranges from 40 to 6000 therefore better to replace with median

```python
df['approx_cost_for_two'].replace(to_replace = 0, value = df['approx_cost_for_two'].median(), inplace=True)
```

**Categorical missing values (location, rest_type , cuisines) can be replaced by the their Mode**

```python
df['location'] = df['location'].fillna(df['location'].value_counts().index[0])
df['cuisines'] = df['cuisines'].fillna(df['cuisines'].value_counts().index[0])
df['rest_type'] = df['rest_type'].fillna(df['rest_type'].value_counts().index[0])
```

# **DATA VISULAIZATIONS**

**Restaurants delivering Online or not**

```python
data = df['online_order'].value_counts()
mylabels = ["YES","NO"]
plt.pie(data, labels = mylabels,autopct='%1.2f%%',explode=[0.1,0],shadow=True)
plt.title("Restaurants delivering Online or not")
plt.legend()
plt.show() 
```

**Restaurants allowing table booking or not**

```python
data = df['book_table'].value_counts()
mylabels = ["NO","YES"]
mycolors = ["hotpink", "b"]
plt.pie(data, labels = mylabels,autopct='%1.2f%%',explode=[0.3,0],shadow=True, colors = mycolors)
plt.title("Restaurants allowing table booking or not")
plt.legend()
plt.show() 
```

**Table booking Rate vs Rate**

```python
avg_rate_with_booking = round(df[df['book_table'] == 'Yes']['rate'].mean(), 2)
avg_rate_without_book = round(df[df['book_table'] == 'No']['rate'].mean(), 2)
dict1 = {'Ratings with Booking':[avg_rate_with_booking],'Ratings without Booking':[avg_rate_without_book]}
df2 =pd.DataFrame.from_dict(dict1)
```

```python
fig , ax = plt.subplots(figsize=(8,4))
sns.barplot(data = df2, palette = 'hls')
for i in ax.containers:
    ax.bar_label(i)
```

**Relation between Location and Rating**

```python
loc_rate = df['rate'].groupby(df['location'],sort=True)
d={}
for i,j in df['location'].value_counts()[:10].to_dict().items():
    d[i]=round(loc_rate.get_group(i).mean(),2)
loc_rate = pd.DataFrame(list(d.items()),columns=['location',"avg_rating"])
```

```python
fig , ax = plt.subplots(figsize=(14,6))
plt.title("Top 10 locations average ratings")
sns.barplot(x="avg_rating",y="location",data=loc_rate.sort_values(by=['avg_rating'],ascending=False))
for i in ax.containers:
    ax.bar_label(i)
```
